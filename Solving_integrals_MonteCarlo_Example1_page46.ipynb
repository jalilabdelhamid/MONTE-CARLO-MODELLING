```python
import numpy as np
from random import random
# Set Number of iterations
num_i = 100

# Set Samples per iterations
N_s = 1000
N=0

# Define function f(x)=?
x=0
y = 3*x**2 + 5

# Define interval [a,b]
a = 1
b = 2

Y = 0
Y_2 = 0

for i in range(1,num_i):
  for j in range(1,N_s):
    # Part for calculate Integral
    x_r = (a + (b-a) * random())
    Y = Y + (3*x_r**2 + 5)*(b-a)
    N+=1
    # Part for calculate Sigma
    Y_2 = Y_2  + (((3*x_r**2 + 5)**2) * ((b-a)**2))

  # Determine integral
  I = (1/N)*Y
  # Detremine Sigma
  sigma = (1/np.sqrt(N)) * np.sqrt((Y_2/N - (Y**2)/N**2))
  print(f" Iteration Number {i} ; I = {I} +/- {sigma}")



```
